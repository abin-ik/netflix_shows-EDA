```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#cleaning the data

df = pd.read_csv("netflix_titles.csv")
df = df.sort_values(by="release_year")
df =df.dropna()
df = df.drop_duplicates()
df
```

```python
plt.figure(figsize=(15,6))
sb.countplot(x = "release_year",data = df,hue="type")
plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```

```python
movies = df[df["type"]== "Movie"]
movies["duration"] = movies["duration"].str.replace("min", "").str.strip().astype(float)
duration_counts = movies["duration"].value_counts().sort_index()

sb.lineplot(x="duration",y =duration_counts,data = movies)
plt.xticks()
```
